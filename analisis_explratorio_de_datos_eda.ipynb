```python
df_español = df.replace(np.nan, '', regex=True)
df_español = df_español[df_español['languages'].str.contains('es')]
df_español
```

```python
df_español.set_index('alpha_3')[['population','area']].plot(kind='bar',rot=65,figsize=(20,10))
```

```python
anomalies=[]
# Funcion ejemplo para deteccion de Outliers

def find_anomalies(data):
    # Seteo limites superior e inferior = media +- 2 veces Desviacion Estandar
    data_std = data.std()
    data_mean = data.mean()
    anomaly_cut_off = data_std * 2
    lower_limit = data_mean - anomaly_cut_off
    upper_limit = data_mean + anomaly_cut_off
    print(lower_limit.iloc[0])
    print(upper_limit.iloc[0])

    # Busqueda de Outliers
    for index, row in data.iterrows():
      outlier = row # Obtener primera columna
      #print(outlier)
      if (outlier.iloc[0] > upper_limit.iloc[0]) or ( outlier.iloc[0] < lower_limit.iloc[0]):
        anomalies.append(index)
    return anomalies

find_anomalies(df_español.set_index('alpha_3')[['population']])


```
